This homework is to predict PM 2.5 density value (a rather large dust particle)
in the air. Features are other meteorological value like temperature, wind, etc.

The data are collected from 豐原(Fong Yuan) observation station
every hour, 24 hours a day, 365 days an year.

The training data consist of the first 20 days of the month.

Testing data are collected randomly from the remaining 10 days,
each datum point represents value of consecutive 10 hours in the period
(And only the first 9 hours are shown)

Our goal is to use the first 9 hours' value as feature to predict 10th hour's PM 2.5


Notes:

NMHC non-methane hydrocarbons

AMB_TEMP ambient temperature 室內溫度

RAIN_FALL NR means no rain

RH relative humidity 相對濕度

THC total hydrocarbon 

WIND_SPEED      m/sec   every hour's last ten minutes avg 風速(以每小時最後10分鐘算術平均)

WIND_DIREC      degress every hour's last ten minutes avg 風向(以每小時最後10分鐘向量平均)

WS_HR           m/sec   whole hour avg 風速小時值(以整個小時向量平均)

WD_HR           degress whole hour avg 風向小時值(以整個小時向量平均)

Time to plan our data research steps.

According to the lecture, we should seperate our training data into train and validate data.
And it's better to do it in many different ways and find the best average performance model.

Our data have 5760 entries. Because of the task is to predict 10th hour value from the past 9 hours, we should set each data point into 10 entries, first 9 for input, last for output comparison.

Therefore we have atmost 5760 - 9 entries available ( We can't predict first 9 entries )
We can randomly decide which entries to predict.
The problem are 
1. How big should the validate dataset be?
2. How many times should we randomly grab those entries?

Anyhow we should try doing it at first.
So let's arbitrarily decicde the number to validate is 500.

There are 18 measurements and 9 time points, which makes 162 parameters.

naive model: y = sigma n=1 to 162: Wn * Xn + b

Loss function = (10th pm2.5 - y) ^ 2

But first we are only giving it a try, so we don't take all the measurements into account,
And I doubt other gases would have any effects on pm2.5.
And wind_direction value is a litte bit tricky, I'll handle them later.

So let's get rid of them:

```python
def result_difference(result_pm25, measurements, weights): # measurements: temp, pm10, pm25, rain_fall, rh, ws, ws_hr,
    return (result_pm25 - (measurements * weights)) ** 2
```

```python
def loss_function(data, weights):
    result = 0
    for datum_point in data:
        answer = datum_point.iloc[9]['PM2.5']
        measurements_series = pd.Series([])
        for index, row in datum_point.iterrows():
            measurements_series.append(row)
        measurements = measurements_series.values
        result += result_difference(answer, measurements, weights)
    return result
```

```python
def partial_diff(b, y, ws_without_b, xs, i=None):
    if i == None:
        return -2 * ( y - ( b + ws_without_b.dot(xs) ) )
    else: 
        return -2 * ( y - ( b + ws_without_b.dot(xs) ) ) * xs[i]

def sigma_adder(data_array, weights, i=None):
    result = 0
    for n in range(len(data_array)):
        b = weights[-1]
        y = data_array[n][-1]
        weights_without_b = weights[:-1]
        xs_without_y = data_array[n][:-1]
        result += partial_diff(b, y ,weights_without_b, xs_without_y, i)
    return result

def calculate_gradient(data_array, weights):
    weights_gradients = []
    # weights
    for i in range(63):
        weights_gradients.append(sigma_adder(data_array, weights, i))        
    # constant b
    weights_gradients.append(sigma_adder(data_array, weights))
    
    return np.array(weights_gradients)
```

```python
weights = np.zeros(64)

weights_gradients = calculate_gradient(first_try_train_data_array, weights)
gradients_history = np.array([weights_gradients])

initial_learning_rates = np.absolute(weights_gradients) * 0.065
```

```python
initial_learning_rates
```

```python
weights_gradients
```

```python

steps = 0

while True:
    steps += 1
    learning_rates = initial_learning_rates / np.sqrt((gradients_history ** 2).sum(0))
    weights = weights - ( learning_rates * weights_gradients )
    weights_gradients = calculate_gradient(first_try_train_data_array, weights)
    gradients_history = np.vstack((gradients_history, weights_gradients))
    if np.all(np.absolute(weights_gradients) < 1):
        break
    
    #if flag_gradient / weights_gradients[0] > 10:
    if steps % 25 == 0:
        print("Step: {}".format(steps))
        print("Gradients: {}".format(weights_gradients))
    
```
